You want to linearly transform points from one coordinate system to another, where the points in the original system are given as two lists, and the new coordinate system maps these points between 0 and 1. This is essentially a normalization problem where we map the original data range to the desired target range (0 to 1).

To achieve this, we can use the following formula for linear transformation:

\[
\text{new\_value} = \frac{(x - \text{min})}{(\text{max} - \text{min})}
\]

Where:
- `x` is the point in the original coordinate system,
- `min` is the minimum value of the original coordinate system,
- `max` is the maximum value of the original coordinate system,
- `new_value` is the transformed point between 0 and 1.

Here's the Python script that implements this transformation:

```python
def transform_coordinates(original_coords, min_val, max_val):
    """
    Transforms a list of coordinates from the range [min_val, max_val] to the range [0, 1].

    Parameters:
    original_coords (list): A list of coordinates in the original coordinate system.
    min_val (float): The minimum value of the original coordinate system.
    max_val (float): The maximum value of the original coordinate system.

    Returns:
    list: A list of transformed coordinates, scaled to the range [0, 1].
    """
    if max_val == min_val:
        raise ValueError("Max and min values cannot be the same, as division by zero will occur.")
    
    # Apply linear transformation
    transformed_coords = [(x - min_val) / (max_val - min_val) for x in original_coords]
    return transformed_coords

# Example usage
# Original coordinates
x_coords = [3, 6, 9, 12]  # x-coordinates in the original system
y_coords = [15, 18, 21, 24]  # y-coordinates in the original system

# Minimum and maximum values for x and y in the original system
min_x, max_x = 0, 15
min_y, max_y = 10, 25

# Transform the coordinates
normalized_x = transform_coordinates(x_coords, min_x, max_x)
normalized_y = transform_coordinates(y_coords, min_y, max_y)

# Print the transformed coordinates
print("Normalized x-coordinates:", normalized_x)
print("Normalized y-coordinates:", normalized_y)
```

### What is an ROC Curve?

ROC stands for Receiver Operating Characteristic. An ROC curve is a graphical plot that illustrates the diagnostic ability of a binary classifier system as its discrimination threshold is varied. It is used to assess how well a model distinguishes between two classes (e.g., positive and negative cases).

### Key Concepts

1. **True Positive Rate (TPR) or Sensitivity:**
   - TPR = True Positives / (True Positives + False Negatives)
   - It measures the proportion of actual positives that are correctly identified by the classifier.

2. **False Positive Rate (FPR):**
   - FPR = False Positives / (False Positives + True Negatives)
   - It measures the proportion of actual negatives that are incorrectly identified as positives by the classifier.

### Steps to Obtain an ROC Curve

1. **Train the Classifier:**
   - First, we train our classifier (e.g., logistic regression, random forest) on the training data. This involves the model learning from the data to distinguish between the two classes.

2. **Predict Probabilities:**
   - After training, the classifier is used to predict the probabilities of the positive class for each instance in the test data. Instead of just predicting the class (positive or negative), we get a probability score (ranging from 0 to 1).

3. **Vary the Threshold:**
   - To create an ROC curve, we vary the decision threshold. The threshold is the value above which we classify an instance as positive. For example, if the threshold is 0.5, any instance with a predicted probability above 0.5 is classified as positive.
   - We evaluate the classifier's performance at multiple thresholds, ranging from 0 to 1.

4. **Calculate TPR and FPR for Each Threshold:**
   - For each threshold, we calculate the TPR and FPR. This gives us a point on the ROC curve.

5. **Plot the ROC Curve:**
   - By plotting the TPR (y-axis) against the FPR (x-axis) for all thresholds, we get the ROC curve. The curve shows the trade-off between sensitivity (true positive rate) and specificity (1 - false positive rate).

### Example with Logistic Regression

Let's go through an example with logistic regression:

1. **Train the Model:**
   - We train a logistic regression model on our training data.

2. **Predict Probabilities:**
   - The model outputs probabilities for each instance in the test set. For instance, an output of 0.7 means the model predicts a 70% chance that the instance is positive.

3. **Evaluate at Multiple Thresholds:**
   - Suppose we evaluate at thresholds 0.1, 0.2, 0.3, ..., 0.9.
   - At each threshold, we classify instances as positive if their predicted probability is above the threshold.

4. **Calculate TPR and FPR:**
   - For threshold 0.1, we calculate TPR and FPR based on the predictions.
   - Repeat this for each threshold.

5. **Plot ROC Curve:**
   - Plot all the TPR and FPR pairs to form the ROC curve.

### Area Under the Curve (AUC)

The AUC is a single number summary of the ROC curve. It represents the probability that a randomly chosen positive instance is ranked higher than a randomly chosen negative instance. An AUC of 1 indicates perfect classification, while an AUC of 0.5 indicates no better than random guessing.
